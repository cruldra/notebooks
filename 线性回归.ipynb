# 线性回归

简单来说,线性回归就是画一条直线,尽量让这条线能够代表数据点的趋势.然后我们科通过这条线可以预测未来的值.

# 核心概念

* 自变量(indenpendent variable): 用来预测的变量,通常用`X`表示
* 因变量(dependent variable): 要预测的变量,通常用`y`表示
* 斜率(slope): 表示自变量的变化对因变量的影响程度
* 截距(intercept): 表示当自变量为0时,因变量的值
* 残差(residual): 表示预测值与真实值之间的差异
* 最小二乘法(least squares method): 通过最小化残差平方和来求解最优的斜率和截距

# 示例1:用线性回归预测房价

```python
# 导入必要的库
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 生成示例数据
# 自变量：房子的面积（平方米）
X = np.array([50, 60, 70, 80, 90]).reshape(-1, 1)  # 使用reshape将数据转换为列向量

# 因变量：房子的价格（万元）
y = np.array([200, 240, 280, 320, 360])

# 创建线性回归模型
model = LinearRegression()

# 训练模型
# 通过fit方法，模型学习面积与价格之间的关系
model.fit(X, y)

# 预测：使用训练好的模型预测新房子的价格
# 假设新房子的面积为85平方米
new_area = np.array([[85]])  # 将新数据转换为与训练数据相同的形状
predicted_price = model.predict(new_area)

# 输出预测结果
print(f"面积为85平方米的房子的预测价格为 {predicted_price[0]} 万元")

# 输出回归系数
# 截距（intercept_）和斜率（coef_）
print("截距:", model.intercept_)
print("斜率:", model.coef_[0])

plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 可视化：绘制数据点和回归线
plt.scatter(X, y, color='blue', label='实际值')  # 绘制实际数据点
plt.plot(X, model.predict(X), color='red', label='回归线')  # 绘制回归线
plt.scatter(new_area, predicted_price, color='green', label='预测点')  # 绘制预测点
plt.xlabel('面积（平方米）')
plt.ylabel('价格（万元）')
plt.legend()
plt.title('面积与房价的线性回归')
plt.show()
```
